```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data['Sex'].replace({"female":0,"male":1},inplace=True)
```

```python
data.head()
```

```python
data.describe()
```

```python

data["Age"] = data["Age"].fillna(data["Age"].mean())
```

```python
data.head()
```

```python
data['Name']
```

```python
data['Ticket'].value_counts
```

```python
def fun2(value):
    if (value == 'S'):
        return 0
    elif (value == 'C'):
        return 1
    elif (value == 'Q'):
        return 2
    else:
        return 0



data["Embarked"] = data["Embarked"].apply(fun2)
```

```python
data.head()
```

```python
sns.countplot(data=data, x='Sex', hue='Survived')
```

```python
# 1. Heatmap to visualize correlation matrix of numerical variables

plt.figure(figsize=(10, 7))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Titanic Dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Set the size of the plot
sns.histplot(data=data,x='Fare', kde=True, binwidth=20)  # Create a histogram with a kernel density estimate
plt.title('Distribution of Ticket Fares on the Titanic')  # Title of the plot
plt.xlabel('Fare')  # Label for the x-axis
plt.ylabel('Number of Passengers')  # Label for the y-axis
plt.show()  # Display the plot
```
