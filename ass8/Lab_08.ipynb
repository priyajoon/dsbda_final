```python
#imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
```

```python
data.sample(5)
```

```python
data.info(), data.isna().sum()
```

```python
sns.countplot(data=data, x='Sex', hue= 'Survived')
```

```python
#sns.heatmap(data.corr(), annot= True, cmap= 'coolwarm', linewidths = 1, linecolor = 'black');
numeric_cols = data.select_dtypes(include=['number']).columns
numeric_data = data[numeric_cols]
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Titanic Dataset')
plt.show()
```

From the above corelation matrix, it is clear that 'Fare' and 'Survived' have a positive corelation.<br>
Meaning higher the cost of the ticket, higher is the chance of survival.<br>

```python
sns.histplot(data = data, x = 'Fare', hue = 'Survived',kde = True);
```

The corelation can be confirmed from the above histplot. <br>
The number of people having cheap tickets was extremely high. However only a few of  them survived.<br>
On the other hand, most of the handful people with expensive tickets survived.<br>
