```python
data={
    'name':pd.Series(['Alice','Bob', 'Charlie', 'David', 'Emma', 'Frank', 'Grace', 'Henry', 'Ivy', 'Jack',
                       'Katie', 'Liam', 'Mia', 'Nate', 'Olivia', 'Peter', 'Quinn', 'Rachel', 'Sam', 'Tyler']),
    'division':pd.Series(['A','B', 'A', 'C', 'B', 'A', 'B', 'C', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A', 'C',
                           'B', 'A', 'C']),
    'english':pd.Series([70, 80, 85, 90, 95, 65, 75, 60, 50, 85, np.nan, 55, 80, 70, 75, 40, 90, 80, 85, 65]),
    'hindi':pd.Series([60, 70, 75, 80, 85, 55, 65, 50, 40, 75, 80, 45, 70, 60, np.nan, 30, 80, 70, 75, 55]),
    'maths':pd.Series([5, 60, 65, 70, 75, 45, 55, 40, 30, 65, 70, 35, 60, 50, 55, 20, 70, 60, np.nan, 45])
}
df=pd.DataFrame(data)
```

```python
mean1=df['english'].mean()
mean2=df['hindi'].mean()
mean3=df['maths'].mean()
```

```python
df['english']=df['english'].fillna(mean1)
df['hindi']=df['hindi'].fillna(mean2)
df['maths']=df['maths'].fillna(mean3)
```

```python
def fun(val):
    return int(math.floor(val))
```

```python
df['english']=df['english'].apply(fun)
df['hindi']=df['hindi'].apply(fun)
df['maths']=df['maths'].apply(fun)
```

iqr method

```python
def detect_outliers(data):
    Q1=data.quantile(0.25)
    Q3=data.quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    outliers=data[(data<lower_bound) | (data>upper_bound)]#check this
    return outliers
```

```python
outliers=detect_outliers(data['maths'])
outliers
```

```python
data_wo_outliers=df[~df['maths'].isin(outliers)]
data_wo_outliers
```

```python
sns.boxplot(data=data_wo_outliers)
```
