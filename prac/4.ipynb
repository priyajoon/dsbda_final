```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('BostonHousing.csv')
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(),annot=True,cmap='coolwarm')
```

```python
train_df=data[['lstat','rm','indus','tax','medv']]
```

```python
sns.pairplot(train_df)
plt.show()
```

```python
x=train_df[['lstat','rm','indus','tax']]
y=train_df['medv']
```

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred[:5]
```

```python
y_pred[:5]
```

```python
mae=mean_absolute_error(y_test,y_pred)
mae
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
sns.regplot(x=y_test,y=y_pred,ci=95,color='blue')
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend(['Regression Line', 'Perfect Predictions'])
```

```python
# Plot the regression line with scatter plot
sns.regplot(x=y_test, y=y_pred, ci=95, scatter_kws={'color': 'blue','alpha': 0.5})


# Add labels and title
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

# Add legend
plt.legend(['Regression Line', 'Perfect Predictions'])

# Show plot
plt.show()
```
