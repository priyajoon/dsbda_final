```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.datasets import load_iris
import warnings
warnings.filterwarnings("ignore")
```

```python
data=sns.load_dataset('iris')
```

```python
data.head()
```

```python
data.info()
```

```python
def convert_num(val):
    if(val=='virginica'):
        return 2
    elif(val=='setosa'):
        return 0
    elif(val=='versicolor'):
        return 1
    else:
        pass

data['species']=data['species'].apply(convert_num)
```

```python
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
sns.histplot(data=data, x='sepal_length', kde=True)
```

```python
sns.histplot(data=data, x='sepal_width',kde=True, color='purple')
```

```python
sns.histplot(data = data, x = 'petal_length', kde= True, color = "green");
```
