```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('titanic')
```

```python
data['sex'].replace({"female":0,"male":1},inplace=True)
```

```python

sns.heatmap(data.corr(numeric_only=True), annot=True)
```

```python
def fun2(value):
    if (value == 'S'):
        return 0
    elif (value == 'C'):
        return 1
    elif (value == 'Q'):
        return 2
    else:
        return 0


data["embarked"] = data["embarked"].apply(fun2)
```

```python


age_mean = data['age'].mean()
data['age'].fillna(age_mean)
```

```python

plt.figure(figsize=(10,7))
sns.boxplot(x='sex', y='age', hue='survived', data=data);
plt.show()
```

```python
grouped = data.groupby(['sex', 'survived'])['age'].describe()

# Print the statistics
print(grouped)
```

```python
print("\nObservations:")
for (sex, survived), group in data.groupby(['sex', 'survived']):
    print(f"\nIn the category {sex}, {'survivors' if survived else 'non-survivors'}:")
    print(f"- Mean Age: {group['age'].mean():.2f}")
    print(f"- Median Age: {group['age'].median():.2f}")
    print(f"- Age Range: {group['age'].min():.2f} to {group['age'].max():.2f}")

    # Determine the opposite survival status
    opposite_survived = 0 if survived else 1

    # Check if the opposite survival status exists in the grouped data
    if (sex, opposite_survived) in grouped.index:
        opposite_median = grouped.loc[(sex, opposite_survived), '50%']
        if group['age'].median() < opposite_median:
            print("  Median age of survivors is lower than non-survivors.")
        else:
            print("  Median age of survivors is higher than non-survivors.")
    else:
        print(f"  No comparable data for {sex} {'non-survivors' if survived else 'survivors'}.")
```
